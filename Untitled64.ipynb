```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms, models
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import time
import warnings
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
```

```python
warnings.filterwarnings("ignore")
```

```python
# Image Transforms
train_transform = transforms.Compose([
    transforms.RandomRotation(10),
    transforms.RandomHorizontalFlip(),
    transforms.Resize(224),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

test_transform = transforms.Compose([
    transforms.Resize(224),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])
```

```python
# Load Dataset
root = '/content/drive/MyDrive/train_test'  # Make sure to include multiple class folders
train_data = datasets.ImageFolder(os.path.join(root, 'Train'), transform=train_transform)
test_data = datasets.ImageFolder(os.path.join(root, 'Test'), transform=test_transform)
```

```python

# Dataloaders
torch.manual_seed(42)
train_loader = DataLoader(train_data, batch_size=10, shuffle=True)
test_loader = DataLoader(test_data, batch_size=10, shuffle=True)

# Class Info
class_names = train_data.classes
num_classes = len(class_names)

print(f"Detected Classes: {class_names}")
print(f"Training Images: {len(train_data)}")
print(f"Testing Images:  {len(test_data)}")
```

```python


# Load Pretrained AlexNet
AlexNetmodel = models.alexnet(pretrained=True)

# Freeze feature extractor
for param in AlexNetmodel.features.parameters():
    param.requires_grad = False

# Modify classifier for MULTICLASS
torch.manual_seed(42)
AlexNetmodel.classifier = nn.Sequential(
    nn.Linear(9216, 1024),
    nn.ReLU(),
    nn.Dropout(0.4),
    nn.Linear(1024, num_classes),        # dynamic class output
    nn.LogSoftmax(dim=1)                 # suitable with NLLLoss
)
```

```python

# Loss & Optimizer
criterion = nn.NLLLoss()
optimizer = torch.optim.Adam(AlexNetmodel.classifier.parameters(), lr=0.001)
```

```python
# Move to device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
AlexNetmodel = AlexNetmodel.to(device)
```

```python
# Training
start_time = time.time()
epochs = 50
max_trn_batch = 800
max_tst_batch = 300

train_losses = []
test_losses = []
train_correct = []
test_correct = []

for i in range(epochs):
    trn_corr = 0
    tst_corr = 0

    AlexNetmodel.train()
    for b, (X_train, y_train) in enumerate(train_loader):
        if b == max_trn_batch:
            break
        b += 1
        X_train, y_train = X_train.to(device), y_train.to(device)

        y_pred = AlexNetmodel(X_train)
        loss = criterion(y_pred, y_train)

        predicted = torch.max(y_pred.data, 1)[1]
        batch_corr = (predicted == y_train).sum()
        trn_corr += batch_corr

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if b % 200 == 0:
            acc = trn_corr.item() * 100 / (10 * b)
            print(f'Epoch: {i}  Batch: {b:4} [{10*b:5}/8000]  Loss: {loss.item():.6f}  Accuracy: {acc:.2f}%')

    train_losses.append(loss)
    train_correct.append(trn_corr)

    AlexNetmodel.eval()
    with torch.no_grad():
        for b, (X_test, y_test) in enumerate(test_loader):
            if b == max_tst_batch:
                break
            X_test, y_test = X_test.to(device), y_test.to(device)
            y_val = AlexNetmodel(X_test)
            predicted = torch.max(y_val.data, 1)[1]
            tst_corr += (predicted == y_test).sum()

    loss = criterion(y_val, y_test)
    test_losses.append(loss)
    test_correct.append(tst_corr)

print(f"\nDuration: {time.time() - start_time:.0f} seconds")
print(f"Final Test Accuracy: {test_correct[-1].item()*100/3000:.2f}%")

# ------------------------------
# Inference on a test image
# ------------------------------
inv_normalize = transforms.Normalize(
    mean=[-0.485/0.229, -0.456/0.224, -0.406/0.225],
    std=[1/0.229, 1/0.224, 1/0.225]
)
```

```python
# Find an index for a snake image
snake_class_index = class_names.index('snakes')
snake_indices = [i for i, (img, label) in enumerate(test_data) if label == snake_class_index]

if snake_indices:
    image_index = snake_indices[0] # Use the first snake image found
else:
    image_index = 0 # Default to the first image if no snake is found
    print("Warning: No snake images found in the test dataset. Displaying the first image instead.")

im = inv_normalize(test_data[image_index][0])
plt.imshow(np.transpose(im.numpy(), (1, 2, 0)))
plt.title(f"Sample Test Image (Predicted: {class_names[AlexNetmodel(test_data[image_index][0].unsqueeze(0).to(device)).argmax(dim=1).item()]})") # Update title to show predicted class
plt.axis("off")
plt.show()

AlexNetmodel.eval()
with torch.no_grad():
    sample_img = test_data[image_index][0].unsqueeze(0).to(device)
    pred_class = AlexNetmodel(sample_img).argmax(dim=1)

print("Predicted class:", class_names[pred_class.item()])
```

```python
# ------------------------------
# Confusion Matrix
# ------------------------------
y_true = []
y_pred = []

AlexNetmodel.eval()
with torch.no_grad():
    for X_test, y_test in test_loader:
        X_test, y_test = X_test.to(device), y_test.to(device)
        outputs = AlexNetmodel(X_test)
        _, preds = torch.max(outputs, 1)
        y_true.extend(y_test.cpu().numpy())
        y_pred.extend(preds.cpu().numpy())

# Compute Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
